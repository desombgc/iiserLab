```python
import numpy as np
import matplotlib.pyplot as plt
y=[0.063,0.070,0.077,0.082,0.084,0.090]
y_square=[i**2 for i in y]
#print(y_square)
y2=[0.003969, 0.0049, 0.005929, 0.0067, 0.0070, 0.0081]
V=[100,120,140,160,180,200]
slope=np.polyfit(V,y2,deg=1)
polynomial=np.poly1d(slope)
#t=polynomial(V)
print(slope)
plt.scatter(V,y2,marker='*',color='red',label='Data Points')
plt.plot(V,polynomial(V),color="blue",label="Fit: y=3.961e-05x")
plt.xlabel('Accelerating Voltage in volt ')
plt.ylabel('Square of Diameter in m$^2$')
plt.title("Diameter$^2$ Vs Accelerating Voltage")
plt.legend(loc = 'best', prop = {'size':10})
plt.grid()
plt.show()



```
